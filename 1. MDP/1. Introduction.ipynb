Before jumping into RL algorithm implementation, it is essential to have some prior knowledge about terminologies and theoretical concepts.

## Markov Property
- It states that the future depends only on the present and is independent of the past history.
- A state $S_t$ is a Markov if and only if
  
  P[$S_{t+1}$ | $S_t$] = P[$S_{t+1}$|$S_1$,$S_2$...$S_n$]

## State Transition Matrix
- The state transition matrix is a 2D matrix that is used to represent the transition probability between states, whereas transition probability is the probability of moving from one to another.
  
- For a Markov state $S$ and successor state $S'$, the state transition probability is
  
  $P_{SS'}$ = P[ $S_{t+1}$ = $S'$ | $S_t$ = S ]

  

## Markov Process or Markov Chain 
- It is a memoryless random process, or simply a sequence of states $S_1$, $S_2$, $S_3$... which follows the Markov property.
- Mathematically, A Markov Process is a tuple $<S, P>$ ,where
   - $S$ is a finite set of states
   - $P$ is a state transition probability $P_{SS'}$

## Markov Reward Process
- A Markov reward process is a Markov process with rewards.
- In a Markov process, there are just a finite set of states and transition probabilities $<S, P>$, there is no reward or value upon reaching another state from any state.
- Mathematically, Markov Reward Process is a tuple $<S, P, R, {\gamma}>$
   - $S$ is a finite set of states
   - $P$ is a state transition probability matrix
   - $R$ is a reward function
     - $R_s$ = $E$[$R_{t+1}$ | $S_t$=$s$] 
   - $\gamma$ is a discount factor $\gamma$ $\in$ $[0, 1]$

**Note:** *We will talk about discount factor $\gamma$ in the upcoming sections.*

- In the above example, rewards are state-based (each state has a fixed reward r(s), independent of the action/next state)

\text{Sleep} & 0
\end{array}
$$



## Markov Decision Process (MDP)
- An MDP is a Markov reward process with decisions.
- In an MRP, rewards are associated with states, meaning the agent gets a certain reward upon reaching a state. On the other hand, in an MDP, the reward is associated with the current state and the action that the agent has performed.
- It is a mathematical framework for modeling sequential decision-making problems where outcomes depend on both random environmental dynamics and an agent’s actions, making it the foundation for formulating and solving reinforcement learning problems.

- Mathematically, An MDP is a tuple $<S, A, P, R, \gamma>$
  - $S$ is a finite set of states
  - $A$ is a finite set of actions
  - $P$ is a state transition probability matrix
  - $R$ is a reward function,
     - $R_s$ = $E$[$R_{t+1}$ | $S_t$=$s$, $A_t$=$a$]
  - $\gamma$ is a discount factor $\gamma$ $\in$ $[0, 1]$

$$
R(s,a) =
\begin{array}{c|cccc}
\text{State} & \text{fb} & \text{study} & \text{pub} & \text{sleep} \\
\hline
\text{Facebook} & -1 & 0 & - & - \\
\text{Class1}   & -1 & 0 & - & - \\
\text{Class2}   & - & 1 & - & 0 \\
\text{Class3}   & - & 2 & -10 & - \\
\text{Pub}      & - & 0 & -10 & - \\
\text{Pass}     & - & - & - & 10 \\
\text{Sleep}    & - & - & - & 0
\end{array}
$$
