# Mean Calculation

## Method 1: non-iterative method
- Collect all the samples and then calculate the average
- drawback: wait until all the samples to be collected, have to recalculate everything if one new item is added in the sample list.

```python
import numpy as np
```

```python
# N samples of integer ranging from 0 to 10
N = 1000
samples = np.random.randint(0, high=10, size=N, dtype='int')
print("Sample length: ", len(samples))
print("first 10 sample values:", samples[:20])
```

```python
mean = samples.sum()/N
print("The non-iterative average of the given samples: ", mean.item())
```

## Method 2: Iterative method
- calculate the average in an incremental and iterative way using this formula:
  
- $mean_{k+1} = mean_k - \frac{1}{k} (mean_k - x_k)$

```python
mean = 0
mean_history = []
for idx, item in enumerate(samples):
    mean = mean - 1/(idx+1) * (mean - item)
    mean_history.append(mean.item())
print("incremental and iterative average of the given sample: ", mean)
```

## Plot mean history

```python
from matplotlib import pyplot as plt 

def plot_line_graph(y_values, x_start=0, x_step=1, title="Line Plot", xlabel="iteration", ylabel="mean"):
    x_values = np.arange(
        x_start,
        x_start + x_step * len(y_values),
        x_step
    )
    plt.figure(figsize=(16, 4))
    plt.plot(x_values, y_values, marker='o', markersize=1)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
plot_line_graph(mean_history)
```

In the above plot, we can clearly see that the mean estimation is not accurate in the beginning due to insufficient samples. As more samples are obtained, the estimate can be improved gradually.
