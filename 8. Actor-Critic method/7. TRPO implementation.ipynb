# 7. Trust Region Policy Optimization implementation (TRPO)

- TRPO is an on-policy algorithm.
- TRPO can be used for environments with either discrete or continuous action spaces.

problem with vanilla policy gradient methods:
- non-stationary input data due to changing policy and output or the reward distribution changes. This makes learning unstable.
- advantage is very random initially

Idea: we can update old policy $\pi_{old}$ to a new policy $\pi_{new}$ such that they are trusted distance apart. Such conservative policy update allows improvement instead of degradation

## Trust Region Method
> A trust region is a concept from optimization (especially used in reinforcement learning and numerical optimization) that controls how big a step you take when updating parameters.

Instead of blindly taking a large step based on your gradient (which might break things), you:

- “Trust” your local approximation of the objective function only within a small region around the current parameters.

So you:
- Build a local model (e.g., linear or quadratic approximation)
- Only update parameters within a safe region (the trust region) where this approximation is reliable

| Method           | Behavior                                      |
| ---------------- | --------------------------------------------- |
| Gradient Descent | Takes step based purely on slope              |
| Trust Region     | Takes step **only if it's within safe limit** |
